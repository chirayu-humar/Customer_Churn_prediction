```python
import warnings
warnings.filterwarnings('ignore')

# import libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, ConfusionMatrixDisplay, accuracy_score, precision_score, recall_score
```

```python
# data import
churn_data = pd.read_csv('churn_data.csv')
cust_data = pd.read_csv('customer_data.csv')
internet_data = pd.read_csv('internet_data.csv')

# data dimensions
print(f'Churn data has {churn_data.shape[0]} rows and {churn_data.shape[1]} columns.')
print(f'Customer data has {cust_data.shape[0]} rows and {cust_data.shape[1]} columns.')
print(f'Internet data has {internet_data.shape[0]} rows and {internet_data.shape[1]} columns.')
```

```python
# columns of the datas
for data in [churn_data, cust_data, internet_data]:
    print(list(data.columns))
```

```python
# data merge
data = pd.merge(churn_data, cust_data, how='inner', on='customerID')
data = pd.merge(data, internet_data, how='inner', on='customerID')

# merged data sample
data.head()
```

```python
# Data types of the columns
data.info()
```

```python
# Analysing the 'TotalCharges' column
data['TotalCharges'].value_counts()
```

```python
# Replacing the empty string columns to nan
data['TotalCharges'] = np.where(data['TotalCharges']==' ', np.nan, data['TotalCharges'])

# Type casting the data
data['TotalCharges'] = data['TotalCharges'].astype('float')
print(data['TotalCharges'].dtype)
```

```python
# Null value check
data.isnull().sum()
```

```python
# analysing the 'TotalCharges' to fix the missing values
sns.boxplot(data['TotalCharges'], orient='h')
plt.show()
```

```python
# the data seems to skewed, so replace the missing value with median
data['TotalCharges'].fillna(data['TotalCharges'].median(), inplace=True)
```

```python
# ploting the target variable
ax = sns.countplot(x = data['Churn'], color='purple')
plt.bar_label(ax.containers[0])
plt.title('TARGET VARIABLE DISTRIBUTION')
plt.show()
```

```python
# Removing the "customerID" column as it doesn't contribute to the analysis
data.drop(['customerID'], axis=1, inplace=True)
```

```python
# data sample
data.head()
```

```python
# One hot encoding the categorical variable
encoded_data = pd.get_dummies(data, drop_first=True, dtype='int')
encoded_data.head()
```

```python
# Multicollinearity Check
plt.figure(figsize=(20,18))
sns.heatmap(encoded_data.corr(), annot=True, fmt='.2f', vmin=-1, vmax=1)
plt.show()
```

```python
# predictor and target data split
x = encoded_data.drop(['Churn_Yes'], axis=1)
y = encoded_data['Churn_Yes']

# train test split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=57)
```

```python
# Data standardisation

scale= StandardScaler()
x_train = scale.fit_transform(x_train)
x_test= scale.transform(x_test)
```

```python
# random Forest
from sklearn.ensemble import RandomForestClassifier

# Initializing and training Random Forest Classifier
rf_classifier = RandomForestClassifier()
rf_classifier.fit(x_train, y_train)
```

```python
# Predictions on testing data
rf_pred = rf_classifier.predict(x_test)

# Evaluating Random Forest Classifier
accuracy = rf_classifier.score(x_test, y_test)
print("Accuracy:", accuracy)
```

```python
# Metrics
cm = ConfusionMatrixDisplay.from_predictions(y_test, rf_pred)
report = classification_report(y_test, rf_pred)
print(report)
```

```python
# Variable Importance
# Getting feature importances from the trained Random Forest model
feature_importances = rf_classifier.feature_importances_

# Mapping feature importance values to corresponding feature names
feature_names = pd.DataFrame(x.columns, columns=['Feature'])
feature_importance_df = pd.DataFrame(feature_importances, columns=['Importance'])
feature_importance_df = pd.concat([feature_names, feature_importance_df], axis=1)

# Sorting features based on importance
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)


feature_importance_df
```

```python
# Plotting Variable Importance
plt.figure(figsize=(10, 6))
plt.barh(feature_importance_df['Feature'], feature_importance_df['Importance'], color='skyblue')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.title('Variable Importance')
plt.gca().invert_yaxis()  # Invert y-axis to display most important features at the top
plt.show()
```

```python
# Select the top 15 features based on their importance
top_features = feature_importance_df['Feature'][:15].tolist()
top_features
```

```python
x = encoded_data[top_features]
y = encoded_data['Churn_Yes']

# train test split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=57)
```

```python
# Balancing the data - synthetic minority over sampling technique
from imblearn.over_sampling import SMOTE

# SMOTE to balance the data
smote = SMOTE(random_state=42)
x_train_balanced, y_train_balanced = smote.fit_resample(x_train, y_train)

# Check the balanced class distribution
print(y_train_balanced.value_counts())
```

```python
# Data standardisation

scale= StandardScaler()
x_train = scale.fit_transform(x_train)
x_test= scale.transform(x_test)
```

```python
# Define the parameter grid for GridSearchCV
param_grid = {'max_depth':[2,3,4,6,8,10,15],
           'max_features':['auto','sqrt','log2'],
           'criterion':['gini','entropy'],
           'n_estimators':[7,11,15,21,27,31]}

# Initialize Random Forest Classifier
rf_classifier = RandomForestClassifier(random_state=42)

# Perform GridSearchCV
grid_search = GridSearchCV(estimator=rf_classifier, param_grid=param_grid, cv=5, scoring='accuracy', n_jobs=-1)
grid_search.fit(x_train_balanced, y_train_balanced)

# Get the best parameters and best score
best_params = grid_search.best_params_
best_score = grid_search.best_score_
print("Best Parameters:", best_params)
print("Best Score:", best_score)
```

```python
y_pred=grid_search.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
# Train the Random Forest model using the best parameters
best_rf_classifier = RandomForestClassifier(n_estimators= 31, max_depth= 15, max_features='sqrt',criterion='gini', random_state=42)
best_rf_classifier.fit(x_train_balanced, y_train_balanced)

# Make predictions on the test set
predictions = best_rf_classifier.predict(x_test)

# Evaluate the model
accuracy = accuracy_score(y_test, predictions)
print("Test Accuracy:", accuracy)

# Classification report
report = classification_report(y_test, predictions)
print(report)

# Confusion Matrix
cm = ConfusionMatrixDisplay.from_predictions(y_test, predictions)
```

```python
from sklearn.tree import DecisionTreeClassifier

# Define the parameter grid for GridSearchCV
param_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [ 10, 20, 30, 40, 50, 60],
    'min_samples_split': [2, 5, 10, 15],
    'min_samples_leaf': [1, 2, 4, 8]
}

# Initialize Decision Tree Classifier
dt_classifier = DecisionTreeClassifier(random_state=42)


# Perform GridSearchCV
grid_search = GridSearchCV(estimator=dt_classifier, param_grid=param_grid, cv=5, scoring='precision', n_jobs=-1)
grid_search.fit(x_train_balanced, y_train_balanced)

# Get the best parameters and best score
best_params = grid_search.best_params_
best_score = grid_search.best_score_
print("Best Parameters:", best_params)
print("Best Score:", best_score)
```

```python
y_pred=grid_search.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
# Make predictions on the test set
predictions = grid_search.predict(x_test)

# Evaluate the model
accuracy = accuracy_score(y_test, predictions)
print("Test Accuracy:", accuracy)

# Classification report
report = classification_report(y_test, predictions)
print(report)

# Confusion Matrix
cm = ConfusionMatrixDisplay.from_predictions(y_test, predictions)
```

```python
from sklearn.neighbors import KNeighborsClassifier

# Initialize KNN classifier
knn_classifier = KNeighborsClassifier()

# Define hyperparameters and their values for GridSearchCV
param_grid = {
    'n_neighbors': [3, 5, 7, 9],  # Number of neighbors
    'metric': ['euclidean', 'manhattan']  # Distance metric
}

# Perform GridSearchCV
grid_search = GridSearchCV(estimator=knn_classifier, param_grid=param_grid, cv=5, scoring='recall', n_jobs=-1)
grid_search.fit(x_train_balanced, y_train_balanced)

# Get the best parameters and best score
best_params = grid_search.best_params_
best_score = grid_search.best_score_
print("Best Parameters:", best_params)
print("Best Score:", best_score)
```

```python
# Make predictions on the test set
predictions = grid_search.predict(x_test)

# Evaluate the model
accuracy = accuracy_score(y_test, predictions)
print("Test Accuracy:", accuracy)

# Classification report
report = classification_report(y_test, predictions)
print(report)

# Confusion Matrix
cm = ConfusionMatrixDisplay.from_predictions(y_test, predictions)
```
